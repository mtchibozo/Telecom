# Data Cleaning

```python
import pandas as pd
```

```python
lines = open("Motorcycledata.txt").readlines()
tab = []
for i in range(len(lines)):
    tmp = lines[i].replace('    ',';')
    tmp = tmp.replace("   ",";")
    tmp = tmp.replace("  ",";")
    tmp = tmp.replace(" ",";")
    row = []
    for string in tmp[:-1].split(';'):
        if string != '':
            row.append(float(string))
    tab.append(row)
tab
```

```python
df = pd.DataFrame()
df[["index","t","Y"]] = pd.DataFrame(tab)
df = df[["index","t","Y"]]
```
